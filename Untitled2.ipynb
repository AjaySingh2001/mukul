```python
df = pd.DataFrame({
    'A': [1, 6, 8, 3, 7],
    'B': [5, 2, 9, 4, 1],
    'C': ['one', 'one', 'two', 'two', 'one']
})

# Set MultiIndex
df = df.set_index(['A', 'C'])

# Sort the MultiIndex
df = df.sort_index()

# Slice MultiIndex DataFrame using .loc
result = df.loc['one':'two']
print(result)
```

```python
df = pd.DataFrame({
    'school_code': ['s001','s002','s003','s001','s002','s004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4'],
    't_id':['t1', 't2', 't3', 't4', 't5', 't6']})


# print(df.loc[2:5])
df.loc[[2,4]]
```

```python
df['weight'].clip(33, 55)
```

```python
# plt.figure(figsize=(10,5))
print(df)
plt.title("Weight of students")
plt.bar(df["name"], df["address"])
```

```python
class_group = df.groupby('class')['weight'].mean()
print(class_group)

plt.figure(figsize=(8,5))
plt.bar(class_group.index, class_group.values)
plt.title("Average Weight by Class")
plt.xlabel("Class")
plt.ylabel("Average Weight")
plt.show()
```

```python
school_count = df['school_code'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(school_count, labels=school_count.index, autopct='%1.1f%%')
plt.title("Students per School")
plt.show()
```
