Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

```python
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 9])
print(s1+s2)
print(s1 - s2)
print(s1 * s2)
print(s1 / s2)
```

```python
s = pd.Series(['100', '200', 'python', '300.12', '400'])
print("Original Data Series:")
print(s)
new_s = pd.concat([s,pd.Series(['500', 'php'])], ignore_index=True)
print("Data Series after adding some data:")
print(new_s)
```

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([2,4,6,8,10])
s3 = s1[~s1.isin(s2)]
print(s3)
```

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([2,4,6,8,10])

# s3 = pd.concat([s1[~s1.isin(s2)], s2[~s2.isin(s1)]])
s3 = pd.Series(np.union1d(s1, s2))
s4 = pd.Series(np.intersect1d(s1,s2))
s5 = s3[~s3.isin(s4)]
print(s5)
```

```python
num_state = np.random.RandomState(50)
num_series = pd.Series(num_state.normal(10,4,20))
print("Minimum, 25th percentile, median, 75th, and maximum of a given series:")
per_series = np.percentile(num_series, q=[0, 25,50,75,100])
print(per_series)
```
